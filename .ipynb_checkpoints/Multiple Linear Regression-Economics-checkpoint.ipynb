```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("economic_index.csv")
df.head()
```

```python
df.info()
```

```python
df.drop(columns=["Unnamed: 0","year","month"],axis=1,inplace=True)
```

```python
df.head()
```

```python
X,y=df[["interest_rate","unemployment_rate"]],df["index_price"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.25,random_state=42)
```

```python
sns.pairplot(df)
```

```python
df.corr()
```

```python
plt.scatter(df['interest_rate'],df['unemployment_rate'],color='black')
```

```python
sns.regplot(x=df['interest_rate'],y=df['index_price'])
```

```python
sns.regplot(x=df['unemployment_rate'],y=df['index_price'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
val_score=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=3)
print(np.mean(val_score))
```

```python
y_pred=regression.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print(score)
#display adjusted R-squared
print(1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
```
